```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Product_Category": ["Electronics", "Clothing", "Groceries", "Furniture", "Toys"],
    "Sales": [120000, 85000, 95000, 40000, 30000]
}

weather_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"],
    "Temperature": [15, 18, 23, 28, 33, 36, 35, 34, 30, 26, 20, 16]
}

expenses = {
    "Category": ["Rent", "Salaries", "Utilities", "Marketing", "Misc"],
    "Expense": [30000, 50000, 10000, 15000, 5000]
}

sales= pd.DataFrame(data)
weather= pd.DataFrame(weather_data)
expense= pd.DataFrame(expenses)

sales.dropna(inplace=True)
weather.dropna(inplace=True)
expense.dropna(inplace=True)

sales["Sales"] = sales["Sales"].astype(float)
weather["Temperature"] = weather["Temperature"].astype(float)

# -----------------------------------
# 3️⃣ Data Analysis
# -----------------------------------
total_sales = sales["Sales"].sum()
avg_temp = weather["Temperature"].mean()
total_expense = expense["Expense"].sum()

# Print summary report
print("----- BUSINESS PERFORMANCE REPORT -----")
print(f"Total Sales: ₹{total_sales:,.0f}")
print(f"Average Monthly Temperature: {avg_temp:.1f}°C")
print(f"Total Expenses: ₹{total_expense:,.0f}")
print("----------------------------------------\n")

plt.figure(figsize=(7, 4))
plt.bar(sales["Product_Category"], sales["Sales"], color='orange', edgecolor='black')
plt.title("Sales by Product Category", fontsize=14, weight='bold')
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Sales (₹)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 4))
plt.plot(weather["Month"], weather["Temperature"], marker='o', color='green', linewidth=2)
plt.title("Monthly Temperature Changes", fontsize=14, weight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(expense["Expense"], labels=expense["Category"], autopct='%1.1f%%',
         wedgeprops={'edgecolor': 'black'})
plt.title("Expense Distribution", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()

summary= pd.DataFrame({
    "Metric": ["Total Sales (₹)", "Average Temperature (°C)", "Total Expenses (₹)"],
    "Value": [total_sales, avg_temp, total_expense]
})

print("SUMMARY TABLE:")
print(summary.to_string(index=False))
```
