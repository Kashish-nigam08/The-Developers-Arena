```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.random.seed(1)
students=pd.DataFrame({
    "Student_id": range(1,31),
    "Gender":np.random.choice(["Male","Female"],size=30),
    "Study hours":np.random.randint(1,10,size=30),
    "Attendance":np.random.randint(60,100,size=30),
    "Marks":np.random.randint(40,100,size=30)
})
print("Sample Data:")
print(students.head(),"\n")
print("Missing values:\n",students.isnull().sum(),"\n")
avg_marks = students["Marks"].mean()
avg_study_hours = students["Study hours"].mean()
top_student = students.loc[students["Marks"].idxmax()]
print("----- STUDENT PERFORMANCE SUMMARY -----")
print("Average Marks:",avg_marks)
print("Average Study Hours:",avg_study_hours)
print(f"Top Scorer: Student {top_student['Student_id']} ({top_student['Marks']} Marks)")
print("----------------------------------------\n")

correlation = students["Study hours"].corr(students["Marks"])
print(f"Correlation between Study Hours and Marks: {correlation:.2f}\n")

plt.figure(figsize=(7, 4))
avg_marks_gender = students.groupby("Gender")["Marks"].mean()
plt.bar(avg_marks_gender.index,avg_marks_gender.values,color=['pink','skyblue'],edgecolor="black")
plt.title("Average Marks by Gender", fontsize=14, weight='bold')
plt.ylabel("Average Marks")
plt.xlabel("Gender")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4))
plt.scatter(students["Study hours"], students["Marks"], color="orange")
plt.title("Relationship Between Study Hours and Marks", fontsize=14, weight='bold')
plt.xlabel("Study Hours per Day")
plt.ylabel("Marks")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(5, 5))
gender_counts = students["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',colors=['skyblue','pink'],wedgeprops={'edgecolor': 'black'})
plt.title("Gender Distribution", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()

print("INSIGHTS:")
print(f"1 The average mark among students is {avg_marks:.1f}, with an average of {avg_study_hours:.1f} study hours per day.")
print("2 The bar chart shows a slight difference in average marks between male and female students.")
print("3 The scatter plot indicates a positive correlation between study hours and marks, that is students who study more tend to score higher.")
print(f"4 Correlation value ({correlation:.2f}) confirms that study time moderately affects performance.")
print(f"5 Student {top_student['Student_id']} achieved the highest score ({top_student['Marks']}).")
```
